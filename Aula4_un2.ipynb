```python
import matplotlib.pyplot as plt

# Dados
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

# Criar um grafico de linha
plt.plot(x, y)

# Adicionar rotulos aos eixos
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# Adicionar um titulo ao gr√°fico
plt.title('Exemplo de Grafico de Linha')

# Exibir grafico
plt.show()
```
